```python
# дані за допомогою методу read_html з таблиці "Коефіцієнт народжуваності в регіонах України (1950—2019)"
import pandas as pd

url = 'https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8'

table = pd.read_html(url, match='Коефіцієнт народжуваності в регіонах України', decimal=',', thousands='.')
df_all = table[0]
df_all
```

```python
# Вивести перші рядки таблиці за допомогою методу head
df_head = table[0].head()
df_head
```

```python
# Визначте кількість рядків та стовпців у датафреймі

print (df_all.shape)

```

```python
# змініть у таблиці значення "—" на значення NaN
non_numeric = df_all.copy()
non_numeric.replace("—", np.nan, inplace=True)
print(non_numeric)
```

```python
# Визначте типи всіх стовпців за допомогою dataframe.dtypes

print (non_numeric.dtypes)
```

```python
# Замініть типи нечислових колонок на числові.
numeric = non_numeric .copy()
numeric = numeric.iloc[:, 1:]
# Замінюємо типи нечислових колонок на числові, використовуючи типи даних в кожній колонці
for column in numeric.columns:
    if numeric[column].dtype == 'object':
        numeric[column] = pd.to_numeric(numeric[column], errors='coerce')
print (numeric.dtypes)
```

```python
# Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)
missing_fraction= df_all.isnull().sum()/len(df_all)
missing_fraction
```

```python
# Видаліть з таблиці дані по всій країні
df_without_ukr= df_all.copy()
index=df_without_ukr.index[-1]
df_without_ukr.drop (index, inplace=True)

df_without_ukr

```

```python
# Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)
import pandas as pd

# Припустимо, що 'non_numeric' - це DataFrame з даними pd.nan
columns =non_numeric.iloc[:,1:]

# Замінити відсутні дані в кожному стовпці на середнє значення цього стовпця
for column in columns:
    columns[column] = pd.to_numeric(columns[column], errors='coerce')
    mean_value = columns[column].mean()
    non_numeric[column].fillna(mean_value, inplace=True)

non_numeric = non_numeric.round(1)

# Вивести змінений DataFrame
print(non_numeric)
```

```python
# Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні
# Роблю копію датасету
table= non_numeric.copy()
#  Видаляю останній рядорк зі значеннями по всій Україні
index=table.index[-1]
table.drop(index, inplace=True)
# Отримую рівень народжуваності в Україні у 2019 році
table['2019'] = pd.to_numeric(table['2019'], errors='coerce')
urk_mean= table['2019'].mean().round(2)
# Отримую список регіонів із рівенем народжуваності вищим ніж urk_mean
brate_ukr_2019 = table[table['2019'] > urk_mean]['Регіон'].tolist()

print(f"Список регіонів із рівнем народжуваності вищим ніж середній по Україні у 2019 році ({urk_mean})")
print(brate_ukr_2019)

```

```python
# Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році
import matplotlib.pyplot as plt

# Припустимо, що 'df_all' - це ваш DataFrame з регіонами, а 'non_numeric' - DataFrame з даними про народжуваність
region = non_numeric.iloc[:, 0]
region_list = region.tolist()

data = non_numeric.iloc[:,-1]


data_list = data.tolist()
data_list = [float(d) for d in data_list]

# plt.figure(figsize=(10, 6))

plt.bar(region_list, data_list, color='skyblue')

plt.xlabel("Region", fontsize="small", color="midnightblue")
plt.ylabel("Birthrate", fontsize="small", color="midnightblue")
plt.title("Birthrate in Ukrainian regions in 2019", fontsize=15)

plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Підготовка даних у форматі "long"
years = non_numeric.columns[1:].tolist()
Volyn = non_numeric.iloc[2, 1:].tolist()
Ukraine = non_numeric.iloc[-1, 1:].tolist()

data = pd.DataFrame({
    'Years': years * 2,
    'Birthrate': Volyn + Ukraine,
    'Region': ['Volyn'] * len(years) + ['Ukraine'] * len(years)
})

# Побудова діаграми віолончелі
plt.figure(figsize=(10, 6))
sns.lineplot(x='Years', y='Birthrate', hue='Region', data=data)
plt.ylim(0, 20)

plt.xlabel('Years', fontsize='small', color='midnightblue')
plt.ylabel('Birthrate', fontsize='small', color='midnightblue')
plt.title('Birthrate Comparison in Volyn and Ukraine', fontsize=15)
plt.legend()

plt.show()
```
