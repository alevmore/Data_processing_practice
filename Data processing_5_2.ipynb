```python
import pandas as pd

# Прочитала і надрукувала вміст файла
table = pd.read_csv("2017_jun_final.csv")
table
```

```python
# Визначте типи всіх стовпців за допомогою dataframe.dtypes
print (table.dtypes)
```

```python
# Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

missing_values = table.isnull().sum()

# Обчислення частки пропусків у кожній колонці
total_rows = len(table)
missing_percentage = ((missing_values / total_rows) * 100).round(2)

# Виведення результату
print("Частка пропусків у кожній колонці:")
missing_percentage
```

```python
# Видалення стовпців з пропущеними значеннями, крім конкретного стовпця
tmp = table[table.columns.difference(['Мова програмування'])].isna().any()
table_drop = table.drop(tmp.index[tmp], axis=1)
table_drop
```

```python
# Знову порахуйте, яка частка пропусків міститься в кожній колонці і переконайтеся, що залишився тільки стовпець "Мова.програмування"
new_missing_values = table_drop.isnull().sum()

# Обчислення частки пропусків у кожній колонці
total_rows = len(table_drop)
new_missing_percentage = ((new_missing_values / total_rows) * 100)

# Виведення результату
new_missing_values
# print("Частка пропусків у кожній колонці:")
# print(new_missing_percentage)
```

```python
# Видаліть усі рядки у вихідній таблиці за допомогою методу dropna
table_rows= table.dropna(axis=0)
table_rows
```

```python
# Визначте новий розмір таблиці за допомогою методу shape
new_shape = table_drop.shape

# Виведення нового розміру таблиці
new_shape
```

```python
# Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python
python_data = table_drop.loc[table_drop['Мова програмування'] == 'Python']

# Виведення вибраних рядків
print(python_data)
```

```python
# Визначте новий розмір таблиці за допомогою методу shape
print(python_data.shape)
```

```python
# Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"
# Групування за стовпчиком "Посада"
grouped_by_position = python_data.groupby('Посада')

# Виведення результатів групування
for position, group in grouped_by_position:

  print('\n', "Посада:", position, '\n')
  print(group)
```

```python
# Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"
aggregated_data = grouped_by_position.agg({'Зарплата на місяць': ['min', 'max']})

# Виведення агрегованих даних
print(aggregated_data)
```

```python
# Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"
def fill_avg_salary(row):
    return row.mean()
# Використання функції fill_avg_salary за допомогою методу apply для створення нового стовпця "avg"
aggregated_data['avg'] = grouped_by_position['Зарплата на місяць'].apply(fill_avg_salary).round(2)


# Виведення таблиці з новим стовпцем "avg"
print(aggregated_data)
```

```python
# Створіть описову статистику за допомогою методу describe для нового стовпчика.
avg_stats = aggregated_data['avg'].describe().round(2)

# Виведення описової статистики
print(avg_stats)
```

```python
# Збереження таблиці у форматі CSV
aggregated_data.to_csv('python_data.csv', index=False)
```
