```python
import pandas as pd
import os
```

```python
# завантажую папку idle і створюю на її основі результуючий датасет df_idle
folder_path = '/content/drive/MyDrive/idle'
files = os.listdir(folder_path)
df_idle = []

for i, file in enumerate(files):
    if file.endswith('.csv'):  # Якщо файл має розширення .csv
        filepath = os.path.join(folder_path, file)  # Формуємо повний шлях до файлу
        df = pd.read_csv(filepath)

        if i == 0 or df_idle == []:  # Якщо це перший файл або список порожній
            df_idle.append(df)
        else:
            # Додаємо файли без першого рядка (рядка із назвами стовпців)
            df_idle.append(df.iloc[1:])

# Повторна індексація усіх датафреймів у списку df_idle
df_idle = pd.concat(df_idle, ignore_index=True)
```

2. Застосовую алгоритм випадкових дерев для класифікації даних утвореного датасету і вивожу метрики якості

```python
# Роблю копію датасету
df= df_concat.copy()
```

```python
# Створюю X i y для машинного навчання
X=df.drop(['class_id'],axis=1)
y=df["class_id"]
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Розділяю датасет на  тренувальну, тестову та валідаційну вибірки
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.33, random_state=42)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler

# Припустимо, що X - це ваш датасет
scaler = StandardScaler()
X_train_norm = scaler.fit_transform(X_train)
X_test_norm = scaler.fit_transform(X_test)
```

```python
X_train_norm [1,:]
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=100, random_state=42)
rfc.fit(X_train_norm,y_train)
y_pred=rfc.predict(X_test_norm)

rfc
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

**!!!Пояснення : мені здається такі високі метрики якості тому, що всі дані взаємоповязані між собою ⚓**

```python
# Вивожу важливість кожної  ознаки у моделі класифікації випадкових дерев rfc

feature_scores = pd.Series(rfc.feature_importances_, index=X_train.columns).sort_values(ascending=False)

feature_scores
```

```python
# Застосовую класс Pipeline в алгоритмі класифікації випадкових дерев
from sklearn.pipeline import Pipeline
```

```python
df_pipe= df_concat.copy()
X_p=df_pipe.drop(['class_id'],axis=1)
y_p=df_pipe["class_id"]
```

```python
X_p.shape, y_p.shape
```

```python
X_p_train,X_p_test,y_p_train,y_p_test=train_test_split(X_p,y_p, test_size=0.33, random_state=42)
```

```python
rf_classifier=RandomForestClassifier()

pipeline_rf=Pipeline(steps=[
    ('scaler', StandardScaler()),
    ("classifier", rf_classifier)
])
pipeline_rf
```

```python
pipeline_rf.fit(X_p_train,y_p_train)
```

```python
predictions=pipeline_rf.predict(X_p_test)
print(classification_report(y_p_test,predictions))
```
